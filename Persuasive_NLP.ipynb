```python
input_str = str(df.iloc[1, 0])+str(df.iloc[1, 1])+str(df.iloc[1, 3])+str(df.iloc[1, 4])
input_str

import string

# turn a doc into clean tokens
def clean_doc(doc):
	# replace '--' with a space ' '
	doc = doc.replace('--', ' ')
	# split into tokens by white space
	tokens = doc.split()
	# remove punctuation from each token
	table = str.maketrans('', '', string.punctuation)
	tokens = [w.translate(table) for w in tokens]
	# remove remaining tokens that are not alphabetic
	tokens = [word for word in tokens if word.isalpha()]
	# make lower case
	tokens = [word.lower() for word in tokens]
	return tokens
```

```python
from nltk.tokenize import wordpunct_tokenize #TreebankWordTokenizer
#tokenizer = TreebankWordTokenizer()
#import nltk
#nltk.download('punkt')
df = df.dropna()
sequences = []
for i in range(1424):
    input_str = str(df.iloc[i]['name'])+' '+str(df.iloc[i]['title'])+' '+str(df.iloc[i]['description'])+' '+str(df.iloc[i]['Material & care'])
    tokens = wordpunct_tokenize(input_str) #clean_doc(input_str)
    line = ' '.join(tokens)
    sequences.append(line)
    data = '\n'.join(sequences)
file = open('inputs.txt', "w")
file.write(data)
file.close()
```
