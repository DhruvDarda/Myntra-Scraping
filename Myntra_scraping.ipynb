```python
import time
from selenium import webdriver
import urllib.request
import pandas as pd
import os
import json
import ssl
ctx = ssl.create_default_context()
ctx.check_hostname = False
ctx.verify_mode = ssl.CERT_NONE
```

```python
def retrieve_links():
    links = []
    driver = webdriver.Chrome('/Users/dhruvdardatrn0213/Downloads/chromedriver*')
    driver.get('https://www.myntra.com/men-ethnic-wear')
    time.sleep(5)
    while(True):
        time.sleep(5)
        for product_base in driver.find_elements_by_class_name('product-base'):
            links.append(product_base.find_element_by_xpath('./a').get_attribute("href"))
        try:
            driver.find_element_by_class_name('pagination-next').click()
        except:
            driver.close()
            driver.quit()
            return links
```

```python
if __name__ == "__main__":
    links = list(set(retrieve_links()))
```

```python
df = pd.DataFrame(columns=['name','title','price','description','Size & fit','Material & care', 'Complete the look'])
metadata = dict.fromkeys(['name','title','price','description','Size & fit','Material & care', 'Complete the look'])

from selenium.common.exceptions import NoSuchElementException

driver = webdriver.Chrome('chromedriver')
for i in range(len(links)):
    driver.get(links[i])
    try:
        metadata['title'] = driver.find_element_by_class_name('pdp-title').get_attribute("innerHTML")
        metadata['name'] = driver.find_element_by_class_name('pdp-name').get_attribute("innerHTML")
        metadata['price'] = driver.find_element_by_class_name('pdp-price').find_element_by_xpath('./strong').get_attribute("innerHTML")
        metadata['description'] = driver.find_element_by_xpath('//*[@id="mountRoot"]/div/div/div/main/div[2]/div[2]/div[8]/div/div[1]/p').text
        metadata['Specifications-Pattern'] = driver.find_element_by_xpath('//*[@id="mountRoot"]/div/div/div/main/div[2]/div[2]/div[8]/div/div[4]/div[1]/div[1]/div[2]').text
        metadata['Specifications-Fabric'] = driver.find_element_by_xpath('//*[@id="mountRoot"]/div/div/div/main/div[2]/div[2]/div[8]/div/div[4]/div[1]/div[2]/div[2]').text
        metadata['Size & fit'] = driver.find_element_by_xpath('//*[@id="mountRoot"]/div/div/div/main/div[2]/div[2]/div[8]/div/div[2]/p/p').text
        metadata['Material & care'] = driver.find_element_by_xpath('//*[@id="mountRoot"]/div/div/div/main/div[2]/div[2]/div[8]/div/div[3]/p').text
        metadata['Complete the look'] = driver.find_element_by_xpath('//*[@id="mountRoot"]/div/div/div/main/div[2]/div[2]/div[8]/div/div[4]/div[2]/div/p/p').text
    except NoSuchElementException:  #spelling error making this code not work as expected
        pass
    print((i*100)/len(links))
    df = df.append(metadata, ignore_index=True)
```
